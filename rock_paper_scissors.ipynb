```python
import tkinter as tk
import random

# Global variables
user_score = 0
computer_score = 0
rounds_to_win = 3
match_over = False

choices = ['Rock', 'Paper', 'Scissors']

def play(choice):
    global user_score, computer_score, match_over

    if match_over:
        return

    user_choice = choice
    comp_choice = random.choice(choices)

    result_text = f"You chose {user_choice}, Computer chose {comp_choice}.\n"

    if user_choice == comp_choice:
        result_text += "It's a tie!"
    elif (
        (user_choice == "Rock" and comp_choice == "Scissors") or
        (user_choice == "Paper" and comp_choice == "Rock") or
        (user_choice == "Scissors" and comp_choice == "Paper")
    ):
        user_score += 1
        result_text += "You win this round!"
    else:
        computer_score += 1
        result_text += "Computer wins this round!"

    # Update scores
    user_score_label.config(text=f"Your Score: {user_score}")
    computer_score_label.config(text=f"Computer Score: {computer_score}")
    result_label.config(text=result_text)

    # Check for match winner
    if user_score == rounds_to_win:
        result_label.config(text=result_text + "\nðŸŽ‰ You won the match! ðŸŽ‰")
        disable_buttons()
    elif computer_score == rounds_to_win:
        result_label.config(text=result_text + "\nðŸ’» Computer won the match!")
        disable_buttons()

def disable_buttons():
    global match_over
    match_over = True
    rock_button.config(state="disabled")
    paper_button.config(state="disabled")
    scissors_button.config(state="disabled")
    play_again_button.config(state="normal")

def reset_game():
    global user_score, computer_score, match_over
    user_score = 0
    computer_score = 0
    match_over = False

    user_score_label.config(text="Your Score: 0")
    computer_score_label.config(text="Computer Score: 0")
    result_label.config(text="")

    rock_button.config(state="normal")
    paper_button.config(state="normal")
    scissors_button.config(state="normal")
    play_again_button.config(state="disabled")

# GUI Setup
root = tk.Tk()
root.title("Rock, Paper, Scissors - by Chitrani")
root.geometry("400x420")
root.config(bg="#e6f7ff")

title = tk.Label(root, text="Rock, Paper, Scissors", font=("Arial", 20, "bold"), bg="#e6f7ff")
title.pack(pady=10)

instruction = tk.Label(root, text="First to 3 wins the match!", font=("Arial", 12), bg="#e6f7ff")
instruction.pack()

result_label = tk.Label(root, text="", font=("Arial", 14), wraplength=350, bg="#e6f7ff", justify="center")
result_label.pack(pady=20)

button_frame = tk.Frame(root, bg="#e6f7ff")
button_frame.pack(pady=10)

rock_button = tk.Button(button_frame, text="Rock", width=10, font=("Arial", 12), command=lambda: play("Rock"))
paper_button = tk.Button(button_frame, text="Paper", width=10, font=("Arial", 12), command=lambda: play("Paper"))
scissors_button = tk.Button(button_frame, text="Scissors", width=10, font=("Arial", 12), command=lambda: play("Scissors"))

rock_button.grid(row=0, column=0, padx=5, pady=5)
paper_button.grid(row=0, column=1, padx=5, pady=5)
scissors_button.grid(row=0, column=2, padx=5, pady=5)

score_frame = tk.Frame(root, bg="#e6f7ff")
score_frame.pack(pady=20)

user_score_label = tk.Label(score_frame, text="Your Score: 0", font=("Arial", 12), bg="#e6f7ff")
computer_score_label = tk.Label(score_frame, text="Computer Score: 0", font=("Arial", 12), bg="#e6f7ff")

user_score_label.pack(side=tk.LEFT, padx=20)
computer_score_label.pack(side=tk.RIGHT, padx=20)

play_again_button = tk.Button(root, text="Play Again", font=("Arial", 12), command=reset_game, state="disabled", bg="orange")
play_again_button.pack(pady=10)

root.mainloop()
```
