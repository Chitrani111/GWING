```python
import tkinter as tk
from tkinter import messagebox
import random

# Global variables
min_val = 1
max_val = 100
secret_number = 0
attempts = 0

# Function to start a new game
def new_game():
    global secret_number, attempts
    secret_number = random.randint(min_val, max_val)
    attempts = 0
    guess_var.set("")
    feedback_label.config(text="")
    attempts_label.config(text="Attempts: 0")
    entry.config(state='normal')
    guess_button.config(state='normal')

# Function to check the guess
def check_guess():
    global attempts
    guess = guess_var.get()
    if not guess.isdigit():
        messagebox.showerror("Invalid input", "Please enter a valid number.")
        return

    guess = int(guess)
    attempts += 1
    attempts_label.config(text=f"Attempts: {attempts}")

    if guess < secret_number:
        feedback_label.config(text="Too low! Try a higher number.")
    elif guess > secret_number:
        feedback_label.config(text="Too high! Try a lower number.")
    else:
        feedback_label.config(text=f"ðŸŽ‰ Correct! The number was {secret_number}")
        entry.config(state='disabled')
        guess_button.config(state='disabled')

# GUI Setup
root = tk.Tk()
root.title("No_Guessing Game by Chitrani")
root.geometry("400x350")
root.config(bg="#f0f8ff")

# Title
title = tk.Label(root, text="Guess the Number!", font=("Arial", 18, "bold"), bg="#f0f8ff")
title.pack(pady=10)

# Instructions
instruction = tk.Label(root, text=f"Guess a number between {min_val} and {max_val}", font=("Arial", 12), bg="#f0f8ff")
instruction.pack()

# Entry box
guess_var = tk.StringVar()
entry = tk.Entry(root, textvariable=guess_var, font=("Arial", 14), justify='center')
entry.pack(pady=10)

# Feedback label
feedback_label = tk.Label(root, text="", font=("Arial", 12), fg="blue", bg="#f0f8ff")
feedback_label.pack(pady=5)

# Buttons
guess_button = tk.Button(root, text="Submit Guess", command=check_guess, font=("Arial", 12), bg="green", fg="white")
guess_button.pack(pady=5)

reset_button = tk.Button(root, text="New Game", command=new_game, font=("Arial", 12), bg="orange", fg="black")
reset_button.pack(pady=5)

# Attempts label
attempts_label = tk.Label(root, text="Attempts: 0", font=("Arial", 12), bg="#f0f8ff")
attempts_label.pack(pady=10)

# Start the game
new_game()

# Start the GUI loop
root.mainloop()
```
