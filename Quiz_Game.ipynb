```python
from tkinter import *
from tkinter import messagebox

# Questions list format:
# ["Question", "Option A", "Option B", "Option C", "Option D", Correct Option Index (1-based)]
questions = [
    ["What house is Harry Potter sorted into at Hogwarts?", "Ravenclaw", "Hufflepuff", "Slytherin", "Gryffindor", 4],
    ["What is the name of Harry's broomstick he receives in his first year?", "Nimbus2000", "Firebolt", "Comet260", "Thunderbolt7", 1],
    ["What creature does Harry talk to in the zoo in The Philosopher's Stone?", "Boa constrictor", "Cobra", "Python", "Anaconda", 1],
    ["Who is revealed to be the Half-Blood Prince?", "Albus Dumbledore", "Severus Snape", "Tom Riddle", "Horace Slughorn", 2],
    ["What is the name of Hagrid's giant spider?", "Aragog", "Shelob", "Norbert", "Fang", 1],
    ["What is the name of the magical object that reveals someone's deepest desire?", "Mirror Of Shadows", "Mirror Of Erised", "Pensieve", "Time-Turner", 2],
    ["Who rescues Harry from the Dursleys at the beginning of The Chamber of Secrets?", "Hagrid", "Ron, Fred & George Weasley", "Dobby", "Sirius Black", 2],
    ["What does Professor Umbridge make Harry do as detention?", "Clean cauldrons", "Sort Files", "Write lines with a cursed quill", "Wash dishes", 3],
    ["Who destroys the final Horcrux, Nagini?", "Harry Potter", "Neville Longbottom", "Ron Weasley", "Hermione Granger", 2],
    ["What form does Luna Lovegood's Patronus take?", "Hare", "Otter", "Fox", "Swan", 1],
]

# Initialize variables
current_question_index = 0
score = 0

# Create root window
root = Tk()
root.title("Harry Potter Quiz by Chitrani")
root.geometry("850x600")
root.config(bg="#f0f8ff")

# Question label
question_var = StringVar()
question_label = Label(root, textvariable=question_var, font=("Arial", 18, "bold"), wraplength=700, bg="#f0f8ff", justify=LEFT)
question_label.pack(pady=20, anchor="w", padx=20)

# Option selection variable
selected_option = IntVar()

# Create radiobuttons for options
option_buttons = []
for i in range(4):
    rb = Radiobutton(root, text="", variable=selected_option, value=i+1, font=("Arial", 16), bg="#f0f8ff", anchor="w", justify=LEFT)
    rb.pack(anchor="w", padx=40, pady=5)
    option_buttons.append(rb)

# Feedback label
feedback_label = Label(root, text="", font=("Arial", 14), bg="#f0f8ff")
feedback_label.pack(pady=10)

# Score label
score_label = Label(root, text="Score: 0", font=("Arial", 14, "bold"), bg="#f0f8ff", fg="green")
score_label.pack()

# Function to load a question
def load_question():
    selected_option.set(0)
    q = questions[current_question_index]
    question_var.set(f"Q{current_question_index+1}: {q[0]}")
    for i in range(4):
        option_buttons[i].config(text=q[i+1])
    feedback_label.config(text="")

# Function to check the answer
def check_answer():
    global score
    selected = selected_option.get()
    if selected == 0:
        messagebox.showwarning("No Selection", "Please select an option before submitting.")
        return

    correct = questions[current_question_index][5]
    if selected == correct:
        feedback_label.config(text="Correct!", fg="green")
        score += 1
        score_label.config(text=f"Score: {score}")
    else:
        correct_answer = questions[current_question_index][correct]
        feedback_label.config(text=f"Wrong! Correct answer: {correct_answer}", fg="red")
    submit_button.config(state=DISABLED)
    next_button.config(state=NORMAL)

# Function to go to the next question
def next_question():
    global current_question_index
    current_question_index += 1
    if current_question_index >= len(questions):
        show_result()
    else:
        load_question()
        submit_button.config(state=NORMAL)
        next_button.config(state=DISABLED)

# Function to show final result
def show_result():
    for widget in root.winfo_children():
        widget.destroy()
    result_text = f"Quiz Completed!\nYour final score is {score} out of {len(questions)}"
    result_label = Label(root, text=result_text, font=("Arial", 20, "bold"), bg="#f0f8ff", fg="blue")
    result_label.pack(pady=100)

# Buttons
submit_button = Button(root, text="Submit", font=("Arial", 14, "bold"), bg="green", fg="white", command=check_answer)
submit_button.pack(pady=10)

next_button = Button(root, text="Next", font=("Arial", 14, "bold"), bg="blue", fg="white", command=next_question, state=DISABLED)
next_button.pack(pady=5)

# Start the quiz
load_question()

root.mainloop()
```
